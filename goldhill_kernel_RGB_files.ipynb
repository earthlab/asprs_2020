```python
fc490 = gpd.read_file(r"C:\Users\jomc9287\Documents\ArcGIS\Default.gdb", driver='FileGDB', layer='proc_490_pts')
fc550 = gpd.read_file(r"C:\Users\jomc9287\Documents\ArcGIS\Default.gdb", driver='FileGDB', layer='proc_550_pts')
fc650 = gpd.read_file(r"C:\Users\jomc9287\Documents\ArcGIS\Default.gdb", driver='FileGDB', layer='proc_650_pts')
fc725 = gpd.read_file(r"C:\Users\jomc9287\Documents\ArcGIS\Default.gdb", driver='FileGDB', layer='proc_725_pts')
fc850 = gpd.read_file(r"C:\Users\jomc9287\Documents\ArcGIS\Default.gdb", driver='FileGDB', layer='proc_850_pts')
fcRGB = gpd.read_file(r"C:\Users\jomc9287\Documents\ArcGIS\Default.gdb", driver='FileGDB', layer='proc_RGB_GeoTaggedPhotosToPo')

fc490['minute'] = [float(dt.split(' ')[1].split(':')[1]) for dt in fc490['DateTime']]
fc550['minute'] = [float(dt.split(' ')[1].split(':')[1]) for dt in fc550['DateTime']]
fc650['minute'] = [float(dt.split(' ')[1].split(':')[1]) for dt in fc650['DateTime']]
fc725['minute'] = [float(dt.split(' ')[1].split(':')[1]) for dt in fc725['DateTime']]
fc850['minute'] = [float(dt.split(' ')[1].split(':')[1]) for dt in fc850['DateTime']]
fcRGB['minute'] = [float(dt.split(' ')[1].split(':')[1]) for dt in fcRGB['DateTime']]

fc850.hist(column='minute', bins=30)
plt.title('850')
plt.show()
```

```python
cams = ['490', '550', '650', '725', '850', 'RGB']
dfs = [fc490, fc550, fc650, fc725, fc850, fcRGB]

fig, axes = plt.subplots(2,3,figsize=(15,10), sharex=True, sharey=True)
i=0
for c, d in zip(cams, dfs):
   
    d.hist(column='minute', bins=30, ax=axes.ravel()[i])
    axes.ravel()[i].set_title(c)
    
    i+=1
    
# plt.xlabel('Minute')
# plt.ylabel('# images')
fig.text(0.5, 0.04, 'Minute', ha='center')
fig.text(0.07, 0.5, '# Images', va='center', rotation='vertical')

plt.suptitle('Image Groupings')
plt.show()
```
